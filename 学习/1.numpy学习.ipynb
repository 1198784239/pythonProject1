'''
某城市一周的最高气温为[28,30,29,31,32,30,29]
计算平均气温，最高气温，最低气温
找出气温超过30°的天气
'''

```python
import numpy as np
temps = np.array([28,30,29,31,32,30,29])
print(temps)
print("平均气温:",'%.2f'%np.mean(temps))
print("最高气温:",np.max(temps))
print("最低气温:",np.min(temps))
print("超过30°的天气",np.cumsum(np.where(temps>30,1,0))[-1])
```

将成绩转化为百分制（假设为10分）

随机数据生成
生成一个（3，4）的随机整数数组，范围[0,10)
计算每列的最大值和每行的最小值
将数组中的所有奇数替换成-1

```python
np.random.seed(0)
arr = np.random.randint(0,10,(3,4))
print(arr)
print('每列的最大值：',np.max(arr,axis=0))#axis=0 是列 1是行
print('每行的最小值：',np.min(arr,axis=1))#axis=0 是列 1是行
print(np.where(arr%2==1,-1,arr))
arr[arr%2==1]=-1
print(arr)
```

生成一个（5，5）的随机数组 范围[0,20)
找出数组中大于10的元素
将所有大于10的元素替换为0

```python
np.random.seed(0)
arr = np.random.randint(0,20,(5,5))
print(arr)
print('大于10的元素',arr[arr>10])
print('大于10的元素替换0',np.where([arr>10],0,arr))
```

统计函数应用
莫公司6个月的销售额为[120,135,110,125,130,140]
计算总销售额，均值，方差
找出销售额最高和最低的月份

```python
arr=np.array([120,135,110,125,130,140])
print('总销售额：',np.sum(arr))
print('均值：',np.mean(arr))
print('方差：',np.var(arr))
print('最高的月份：',np.argmax(arr)+1) #arg找到索引
print('最高的月份：',np.argmin(arr)+1)
```

8.数组拼接
A=[1,2,3] B=[4,5,6]
将A和B水平拼接成一个新的数组
将A和B垂直拼接成一个新的数组

```python
A=np.array([1,2,3])
B=np.array([4,5,6])
C=np.concatenate([A,B])
print(C)
print(np.reshape(C,(2,3)))
```

唯一值和排序
数组[2,1,2,3,1,4,3]
找出数组中的唯一值并排序
计算每一个唯一值出现的次数

```python
arr=np.array([2,1,2,3,1,4,3])
u_arr,count=np.unique(arr,return_counts=True)
print(u_arr)
print(count)
d= []
for i in range(len(u_arr)):
    d = d + [len(arr[arr==u_arr[i]])]
print(d)
```

综合练习
销售额：[20,25,22,30,28]
成本：[15,18,16,22,20]
计算每天的利润
计算利润的平均值和标准差
找出利润最高的天数

```python
moneny = np.array([20,25,22,30,28])
cb = np.array([15,18,16,22,20])
cost  = moneny-cb
print('利润:',cost)
print('平均:',np.mean(cost))
print('标准差:',np.std(cost))
print('最高的天数:',len(cost[cost==np.max(cost)]))
```
