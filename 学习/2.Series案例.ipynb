tip1:创建一个包含10名学生数学成绩，成绩范围50-100之间计算平均值，最高分，最低分，找出高于平均分的人数

```python
import numpy as np
import pandas as pd

np.random.seed(0)
score =pd.Series(np.random.randint(50,101,10),index=['学生'+str(i) for i in range(1,11)]) 
print(score)
print('最高分',max(score))
print('最低分',min(score))
print('平均分',score.mean())
mean  = score.mean()
print('高于平均分人数',len(score[score > mean]))
print('高于平均分人数',score[score > mean].count())
```

tip2:温度数据分析
找出温度超过30°的天气
计算平均温度
温度从低到高排序
找出温差变化最大的天数

```python
temper = pd.Series([28,31,29,32,30,27,33],index=['周一','周二','周三','周四','周五','周六','周日'])
print('平均温度',temper.mean())
print('排序',temper.sort_values())
t3=temper.diff().abs()  #计算数组Series变化值
t4 = t3.sort_values(ascending=False).index[:2].tolist()
print("最高的温度",*(t4))  #*()去掉方括号
```

tip3:股票价格分析
收益率（重要 pct_change())
最高值最低值日期

```python
np.random.seed(0)
data = pd.date_range('20180101', periods=6)
prices = pd.Series(np.random.randint(100,120,6),index=data)
syl =prices.pct_change()
print('收益率:',syl)
print('收益率最高的:',syl.idxmax())
print('波动率:',syl.std())
```

tip4:销售数据分析
计算季度平均销量（3个月为一个季度）
找出销量最高的月份
计算月环比增长额
找出连续增长超过两个月的月份

```python
sale = pd.Series([120,135,145,160,155,170,180,175,190,200,210,220],index=pd.date_range('2022-01-01', periods=12,freq='MS'))
print(sale)
#季度的平均销量
sale.resample('QS').mean() #重新采样
print('销量最高的月份:',sale.idxmax())
print('月环比增长率:',sale.pct_change())
```

```python
# 找出连续增长超过两个月的月份
a = sale.pct_change()
b = a>0
b[b.rolling(3).sum()==3].keys().tolist()
```

tip5:每小时销售数据分析
按天重采样计算每日总销售量
计算营业时间8：00-22：00 和非营业时间的销售比例
找出销售额最高的3个小时

```python
np.random.seed(42)
hourly_sales = pd.Series(np.random.randint(0,100,24),index=pd.date_range('2025-01-01', periods=24,freq='h'))
print(hourly_sales)
```

```python
# 按天重新采样计算每日总销售
day = hourly_sales.resample('D').sum()
```

```python
# 按天重新采样计算每日总销售
day = hourly_sales.resample('D').sum()
# 计算营业时间8：00-22：00 和非营业时间的销售比例
hourly_sales.between_time('8:00','22:00')  #筛选一段时间的Series
hs = hourly_sales[(hourly_sales.index.hour>=8)&(hourly_sales.index.hour<=22)] #布尔索引 一个条件一个条件写，最后加Serious
n_b=~hs
print(n_b)
#取销售额最高的三个小时
print(hourly_sales.nlargest(3).values)
print(hourly_sales.nlargest(3).keys())
```
